```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
```

```python
#reading csv and making dataframe
weather_data = "output_data/weather_info.csv"
weather_data_df = pd.read_csv(weather_data)
weather_data_df
```

```python
#Humidity Heatmap

# Configure gmaps with API key
gmaps.configure(api_key=g_key)
```

```python
locations =weather_data_df[["Lat", "Long"]].astype(float)
#type(weather_data_df["Lat"][1])
weather_data_df = weather_data_df.dropna()
weather_data_df
```

```python
fig = gmaps.figure()
heat_layer = gmaps.heatmap_layer(locations, weights=weather_data_df["Humidity"], dissipating=False, max_intensity=100,
                                 point_radius = 1)
fig.add_layer(heat_layer)
fig
#fig.save('output_data/Heat_Map1.png')
```

```python
#finding the ideal weather conditions
ideal_weather = weather_data_df.loc[weather_data_df["Max Temp"]>=70]
ideal_weather2 = ideal_weather.loc[ideal_weather["Max Temp"]<80]
ideal_weather3 = ideal_weather2.loc[ideal_weather2["Wind Speed"]<10]
ideal_weather4 = ideal_weather3.loc[ideal_weather3["Cloudiness"]==0]
ideal_weather4
```

```python
#Putting a blank column for the api call
ideal_weather4['Closest Hotel'] = ""
locations2 =ideal_weather4[["Lat", "Long"]].astype(float)
ideal_weather4
```

```python
locations2
```

```python
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
params = {
    "location": locations2["Lat"] + locations2["Long"],
    "rankby": "distance",
    "type": "hotel",
    "key": g_key,
}
# use iterrows to iterate through pandas dataframe
for index, row in ideal_weather4.iterrows():

    # get hotel from df
    hotel_type = row['City']

    # add keyword to params dict
    params['keyword'] = hotel_type

    # assemble url and make API request
    print(f"Retrieving Results for Index {index}: {hotel_type}.")
    response = requests.get(base_url, params=params).json()
    
    # extract results
    results = response['results']
    
    try:
        print(f"Closest {hotel_type} hotel is {results[0]['name']}.")
        
        ideal_weather4.loc[index, 'Closet Hotel'] = results[0]['Closet Hotel']
        #types_df.loc[index, 'address'] = results[0]['vicinity']
        #types_df.loc[index, 'price_level'] = results[0]['price_level']
        #types_df.loc[index, 'rating'] = results[0]['rating']
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
    print("------------") 
```
