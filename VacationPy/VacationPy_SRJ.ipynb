```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
```

```python
#reading csv and making dataframe
weather_data = "output_data/weather_info.csv"
weather_data_df = pd.read_csv(weather_data)
weather_data_df
```

```python
#Humidity Heatmap

# Configure gmaps with API key
gmaps.configure(api_key=g_key)
```

```python
locations =weather_data_df[["Lat", "Long"]].astype(float)
#type(weather_data_df["Lat"][1])
weather_data_df = weather_data_df.dropna()
weather_data_df
```

```python
fig = gmaps.figure()
heat_layer = gmaps.heatmap_layer(locations, weights=weather_data_df["Humidity"], dissipating=False, max_intensity=100,
                                 point_radius = 1)
fig.add_layer(heat_layer)
fig
#fig.save('output_data/Heat_Map1.png')
```

```python
#finding the ideal weather conditions
ideal_weather = weather_data_df.loc[weather_data_df["Max Temp"]>=70]
ideal_weather2 = ideal_weather.loc[ideal_weather["Max Temp"]<80]
ideal_weather3 = ideal_weather2.loc[ideal_weather2["Wind Speed"]<10]
ideal_weather4 = ideal_weather3.loc[ideal_weather3["Cloudiness"]==0]
ideal_weather4
```

```python
ideal_weather4 = ideal_weather4.reset_index()
ideal_weather4
```

```python
hotel=[]
for unit in range(len(ideal_weather4)):
    lat = ideal_weather4.loc[unit]['Lat']
    lng = ideal_weather4.loc[unit]['Long']
    params = {
        "location": f"{lat},{lng}",
        "radius": 5000,
        "types" : "hotel",
        "key": g_key
    }
    googleurl = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    request = requests.get(googleurl, params=params)
    response= request.json()
    response
    try:
        hotel.append(response['results'][0]['name'])
    except:
        print(f"hotel not found")
ideal_weather4["Hotel Name"] = hotel
ideal_weather4
```

```python
city_l = ideal_weather4["City"].tolist()
city_l
```

```python
# Create a map using state centroid coordinates to set markers
marker_locations = ideal_weather4[['Lat', 'Long']]

list=[f"Name: {ideal_weather4.loc[x]['City']} <br/>Hotel: {ideal_weather4.loc[x]['Hotel Name']}<br/>Country: {ideal_weather4.loc[x]['Country']}" for x in range(len(ideal_weather4))]

# Create a marker_layer using the poverty list to fill the info box
fig = gmaps.figure()
content = [f"Name: {city}" for city in city_l]  
markers = gmaps.marker_layer(marker_locations,
    info_box_content=list)
fig.add_layer(markers)
fig
```
